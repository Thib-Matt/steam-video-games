```python
def split_link(x):
    (head, sep, tail) = x.partition('?')
    return head 
```

```python
# get the video game's ID from the link
def get_game_id(a):
    b = a.partition('/')
    c = b[2].partition('/')
    d = c[2].partition('/')
    e = d[2].partition('/')
    (game_id, sep, name) = e[2].partition('/')
    return game_id
```

```python
# get the video game's name from the link
def get_game_name(a):
    b = a.partition('/')
    c = b[2].partition('/')
    d = c[2].partition('/')
    e = d[2].partition('/')
    f = e[2].partition('/')
    (name, sep1, sep2) = f[2].partition('/')
    return name
```

## Scraping video game webpage

```python
#Scrapping video game webpage
def scrape_content3(soup):
    #get the name
    try:
        name = soup.find('div', class_ = 'apphub_AppName').get_text()
    except:
        name = np.nan        
    #get the reviews
    try:
        user_review = soup.find('span', itemprop="description" ).get_text()
    except:
        user_review = np.nan
    try:
        a = soup.findAll('div', {'class': "user_reviews"})
        b = a[0].findAll('div', {'class': "summary column"})
        nb_reviews = b[1].find('span', {'class':"responsive_hidden"}).get_text().strip()     
    except:
        nb_reviews = np.nan
    #get the release date
    try:
        release_date = soup.find('div', class_ = 'date').get_text()
    except:
        release_date = np.nan
    #get list of available languages
    try:
        language_get = soup.find('table', class_ = 'game_language_options').find_all('td', class_ = 'ellipsis')
        l_language_get = []
        for x in language_get:
            l_language_get.append(x.get_text().strip())
    except: 
        l_language_get = np.nan
    #get list of tags
    try:
        tag_get = soup.find('div', class_ = 'glance_tags popular_tags').find_all('a', class_ = 'app_tag')
        tags = []
        for x in tag_get:
            tags.append(x.get_text().strip())
    except: 
        tags = np.nan
    #get the price
    try:
        tag_price = soup.find('div', {'class': "game_purchase_price price"}).get_text().strip()
    except:
        tag_price = np.nan
    #get PEGI index
    try:
        tag_pegi = soup.findAll('div', {'class': 'block responsive_apppage_details_right'})
        pegi_link = tag_pegi[1].find('td').find('img').get('src')
        pegi_link = pegi_link.split('/')
        pegi = pegi_link[-1]
        pegi = pegi.partition('.')
        pegi = pegi[0]
    except:
        pegi = np.nan
    #get names fo publisher and developper
    try: 
        tag_dev_pub = soup.find('div', {'class': "user_reviews"}).findAll('div', {'class': "dev_row"})
        tag_dev = tag_dev_pub[0].find('div', {'class': 'summary column'}).get_text().strip()
    except:
        tag_dev_pub = np.nan
        tag_dev_pub = soup.find('div', {'class': "user_reviews"}).findAll('div', {'class': "dev_row"})
        tag_dev = np.nan
    try: 
        tag_dev_pub = soup.find('div', {'class': "user_reviews"}).findAll('div', {'class': "dev_row"})
        tag_pub = tag_dev_pub[1].find('div', {'class': 'summary column'}).get_text().strip()
    except:
        tag_dev_pub
        tag_dev_pub = soup.find('div', {'class': "user_reviews"}).findAll('div', {'class': "dev_row"})
        tag_pub = np.nan  
    #get system requirements
    try:
        sr = soup.find('div', {'class': "game_area_sys_req sysreq_content active"}).get_text().strip()
        (head, sep, tail)= sr.partition('\n')
        system_req = head
    except:
        sr = np.nan
        system_req = np.nan
      
    return name, user_review, nb_reviews, release_date, l_language_get, tags, tag_price, pegi, tag_dev, tag_pub, system_req
```

```python
#create an empty dataframe containing all the scrapped information from each video game web page
columns = [
    'game_id',
    'Name',
    'year',
    'type',
    'User_review',
    'Number of reviews',
    'Release_date',
    'Language',
    'Tags',
    'Price',
    'PEGI',
    'Developer',
    'Publisher',
    'System Requirements']
df2 = pd.DataFrame(columns = columns)
df2
```
